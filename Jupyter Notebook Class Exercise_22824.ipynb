```python
file = open ('nyc_data.txt', 'r')
f = file.readlines()
temps= []
degree_symbol="\degree"

for line in f:
        temps.append(line[:-1])


import math
period=365
y_offset=55
amplitude=25
time_offset=200
yt=[]
for x in range(1,366):
    y_time=y_offset+amplitude*math.cos(2*math.pi*(x-time_offset)/period)
    yt.append(y_time)

import math
temps_float=[]
for i in range(1,366):
    temp=float(temps[i-1])
    temps_float.append(temp)
residuals=[]
for i in range(1,366):
    res=yt[i-1]-temps_float[i-1]
    residuals.append(res)

print(f"{'Days':^10}{'Temp(F)':^10}{'Pre-fit Model Values':^10}{'Residual Times':^20}")


for x in range(365):
    print(f"{x+1:^10}{temps[x]:^10}{yt[x]:^10}{residuals[x]:^30}")
```
